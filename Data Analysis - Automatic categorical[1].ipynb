```python
import pandas as pd
```

```python
dataframe = pd.read_csv("googleplaystore.csv").dropna() 
```

```python
dataframe
```

```python
for i in dataframe['Category'].unique():
    print(i) 
```

# Q1 Total number of apps in each category

```python
categories= {}

for name in dataframe['Category'].unique():
    ct = 0
    for i in dataframe['Category']:
        if i == name:
            ct += 1
            categories[name] = ct

print(categories)
```

# Q2 Total number of apps in each Type

```python
types = {}

for name in dataframe["Type"].unique():
    ct = 0
    for i in dataframe["Type"]:
        if i == name:
            ct += 1
    types[name] = ct

print(types)
```

# Q3 Total number of apps in Content Rating

```python
content_rating = {}

for name in dataframe["Content Rating"].unique():
    ct = 0
    for i in dataframe["Content Rating"]:
        if i == name:
            ct += 1
    content_rating[name] = ct

print(content_rating)
```
